```
df = pd.read_csv('/content/drive/My Drive/Data for Python/cleaned_311_data_hw3.csv', header='infer')
```

```
new  = df.groupby(['complaint_type', 'descriptor']).size().reset_index(name='count')
new
```

```
def sf(row):
  if 'Graffiti'  in row.complaint_type:
    return True 
  elif 'Graffiti' in row.descriptor:
    return True
  else:
    return False
```

```
type(sf)
```

```
new['graffiti'] = new.apply(sf, axis=1)
```
